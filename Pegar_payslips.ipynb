This Python program's objective is to automatically extract PDF attachments from emails received with the "maeve" command and store them in a specified local folder. The imaplib package, which enables the retrieval and management of emails via the IMAP protocol, is used by the software to connect to your email server in order to accomplish this. Once connected, it looks through your mailbox for emails that were sent by "maeve" in particular. It recognizes these emails and parses them to extract any attachments that are in PDF format. Then, these attachments are saved into the designated "Payslips_2" directory, which is immediately created if it doesn't already exist. By appending a sequential numbering scheme to the filename, each attachment is given a unique name to prevent overwriting already-existing files.

```python
#working for excel 
# emails maevecarley@me.com, maevecarley@icloud.com
lista_emails_maeve=['maevecarley@me.com', 'maevecarley@icloud.com']
lista_pdfs =[]
for emails in lista_emails_maeve:
    lista_emails = meu_email.fetch(AND(from_=emails)) #cria a lista de emails

    total_relatorio_emails =0
    total_emails =0
    file_path = pathlib.Path.cwd()
    file_path = file_path / ("Payslips_2")
    
    for email in lista_emails: #percorre cada email dentro da lista de emails
        total_emails +=1 # conta quantos emails tem
        if len(email.attachments) > 0: #descobre se existe pelo menos 1 anexo
            for anexo in email.attachments: # passa por todos os anexos dentro do email
                if "Payslip" in anexo.filename: #descobre se existe arquivo que contenha payslip no nome
                    print(anexo.content_type)
                    pagamento = anexo.payload
                    
                    if anexo.filename in lista_pdfs:
                        new_file_name = file_path / (anexo.filename[:-4]+ "_1" +".pdf")
                        with open(new_file_name,"wb") as arquivo_excel:
                            arquivo_excel.write(pagamento)
                            print('arquivo salvo')
                            total_relatorio_emails +=1 #conta quantos emails possui payslip
                        lista_pdfs.append(anexo.filename)    
                        
                    else:
                        with open(file_path / anexo.filename ,"wb") as arquivo_excel:
                                arquivo_excel.write(pagamento)
                                print('arquivo salvo')
                                total_relatorio_emails +=1 #conta quantos emails possui payslip  
                    
                    lista_pdfs.append(anexo.filename)
                    print(lista_pdfs)
                    

    print(total_emails )
    print(total_relatorio_emails ) 
```
